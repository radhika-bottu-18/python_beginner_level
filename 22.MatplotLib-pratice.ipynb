```python
x= np.arange(1,5)
fig,ax= plt.subplots()
ax.plot(x,x/5,label='Normal')
ax.plot(x,x/3,label='Fast')
ax.plot(x,x/2,label='Slow')
ax.legend()
plt.show()
```

```python
x1= np.arange(1,10)
plt.plot(x1,x1**2,'r--',label='Normal')
plt.plot(x1,x1**3,'b^',label='Medium')
plt.legend(loc=1)
plt.show()
```

```python
y = np.arange(1,10)
fig,axes = plt.subplots(2)
axes[0].plot(y,y**2,label='Leg1')
axes[0].plot(y,y/2,label='leg2')
axes[0].set_xlabel('Xlabel')
axes[0].set_ylabel('Ylabel')
axes[0].legend(loc=2)

axes[1].plot(y,np.sin(y),label='leg3')
axes[1].plot(y,np.cos(y),label='leg45')
axes[1].legend(loc=1)
plt.show()
```
